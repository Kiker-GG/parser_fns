```python
inn = np.empty(900, dtype=np.int64)
for page in range(1, 10):
    URL = f"https://companies.rbc.ru/okved/41.20/{page}/"
    headers = {'User-Agent': 'Mozilla/5.0'}
    session = requests.Session() 
    res = session.get(URL, headers=headers)
    soup = BeautifulSoup(res.text, 'html.parser')
    class_name = 'company-card__info'
    p = soup.find_all('p', class_=class_name)
    
    j = 0
    for i in range(0, len(p)):
        elem = p[i]
        span = elem.find('span')
        if span and 'ИНН' in span.get_text():
            inn[(page - 1) * 100 + j] = elem.get_text().split(':')[1]
            j += 1
            
    sleep(0.5)
    
def app_data(key, subkey, data, my_dict):
    if not(key in my_dict):
        my_dict[key] = {}
    my_dict[key][subkey] = data
    
EBITDA = {}
R = {}
C = {}

# inn
```

```python
for i in inn:
    URL = f"https://bo.nalog.ru/advanced-search/organizations/search?query={i}&page=0"
    session = requests.Session()
    try:
        res = session.get(URL, headers=headers).json()
    except Exception as e: 
        print(f"Ошибка: {e}")
        continue
        
    # print(res['content'][0]['id'])
    if (len(res['content']) == 0):
        continue
    # print(res)

    C[i] = res['content'][0]['shortName']
    
    id = res['content'][0]['id']
    URL = f"https://bo.nalog.ru/nbo/organizations/{id}/bfo/"
    res = session.get(URL, headers=headers).json()
    
    for j in range(0, len(res)):
        # net_profit, income_tax, interest_on_loans = 0
        try: 
            net_profit = res[j]['correction']['financialResult']['current2400']
        except Exception as e:
            # print(f"Ошибка: {e}")
            net_profit = 0
        try:
            income_tax = res[j]['correction']['financialResult']['current2410']
        except Exception as e:
            # print(f"Ошибка: {e}")
            income_tax = 0
        try: 
            interest_on_loans = res[j]['correction']['financialResult']['current2330']
        except Exception as e:
            # print(f"Ошибка: {e}")
            interest_on_loans = 0

        #print(res)
        #print('За ', res[j]['period'], 'год ', net_profit, ' + ', income_tax, ' + ', interest_on_loans, ' = ', net_profit + income_tax + interest_on_loans)
        
        # try:
        #     EBITDA[res[0]['organizationInfo']['inn']].append(net_profit + income_tax + interest_on_loans)
        # except Exception as e:
        #     # print(f"Ошибка: {e}")
        #     EBITDA[res[0]['organizationInfo']['inn']] = [net_profit + income_tax + interest_on_loans]

        app_data(res[0]['organizationInfo']['inn'], res[j]['period'], net_profit + income_tax + interest_on_loans, EBITDA)

    try:
        R[res[0]['organizationInfo']['inn']] = int(res[0]['correction']['financialResult']['current2400'] * 100 / res[2]['correction']['financialResult']['current2400']) - 100
        # print(res[0]['correction']['financialResult'])
        # print(res[2]['correction']['financialResult'])
        # print(res[0]['correction']['financialResult']['current2400'] * 100 / res[2]['correction']['financialResult']['current2400'])
        

    except Exception as e:
        R[res[0]['organizationInfo']['inn']] = 0
    
    sleep(0.25)
 
# k = 0
# c = 0
# count = np.zeros(len(EBITDA))

y_2023, y_2022, y_2021, y_2020 = (0, 0, 0, 0)
count_y2023, count_y2022, count_y2021, count_y2020 = (0, 0, 0, 0) 

for i in EBITDA:
    for j in EBITDA[i]:
        match j:
            case '2023':
                y_2023 += EBITDA[i][j]
                count_y2023 += 1
            case '2022':
                y_2022 += EBITDA[i][j]
                count_y2022 += 1
            case '2021':
                y_2021 += EBITDA[i][j]
                count_y2021 += 1
            case '2020':
                y_2020 += EBITDA[i][j]
                count_y2020 += 1
        
        #print(EBITDA[i][j], "    ", j, "    ", i)

# c += k

# print(count)
avg_y2023 = y_2023 / count_y2023
avg_y2022 = y_2022 / count_y2022
avg_y2021 = y_2021 / count_y2021
avg_y2020 = y_2020 / count_y2020

#График 

x = [avg_y2020, avg_y2021, avg_y2022, avg_y2023]
y = [2020, 2021, 2022, 2023]

plt.plot(x, y)
plt.title('Средний показатель EBITDA по годам')
plt.yticks(y)

plt.show()

# print(EBITDA)
# print(C)
```

```python
data = {}

for i in R:
    if R[i] == 0 or abs(R[i]) > 10000:
        continue
        
    if not(R[i] in data):
        data[R[i]] = 0
        
    data[R[i]] += 1

# print(data)

#График 

x = list(data.values())
y = list(data.keys())

d1 = np.arange(-20000, 15500, 500)
d2 = np.arange(0, 25, 1)

plt.figure(figsize=(15,15))
plt.scatter(x, y, s=10)
plt.title('График')  
plt.grid()
plt.xticks(d2)
plt.yticks(d1)

plt.show()
```
